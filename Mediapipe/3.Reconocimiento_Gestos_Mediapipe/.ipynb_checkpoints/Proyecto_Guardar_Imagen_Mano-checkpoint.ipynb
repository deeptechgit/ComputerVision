## Proyecto Capturar la imagen de la mano para crear datasets

```python
## Importacion de librerias
import cv2 as cv
import mediapipe as mp

## referenciando metodos necesarios
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands

# configurando parametros de deteccion
hands = mp_hands.Hands(max_num_hands=2, # numero maximo de manos a detectar
                       min_detection_confidence=0.8,  # porcentaje de precisi칩n en la detecci칩n
                       min_tracking_confidence=0.8  # porcentaje de precisi칩n en el tracking
                      )
# creacion de ventanas para mostrar los resultados
cv.namedWindow("MediaPipe Hands", cv.WINDOW_NORMAL)
cont =0  # contador para el numero de imagen 
xlist=[] # lista para las cordenadas en x
ylist=[] # lista para almacenar las cordenadas en y
## *****************************************************************************
## Funcion para dibujar los puntos y sus conexiones
def hand_detect(img):
    
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB) # cambiamos el espacio de color a rbg porque mediapipe acepta rgb
    result = hands.process(img)  # realizamos la detecion
    h,w,c = img.shape # obtenemos las dimensiones de la imagen 
    coord=[]  # lista para almacenar las cordenadas de con la posicion de la imagen 
    if result.multi_hand_landmarks: # si nuestro retorno de el metodo process no esta vacio
        
    
        for hand_landmarks in result.multi_hand_landmarks: # interamos en el numeor de manos detectada
            # dibujamos las conexiones
            mp_drawing.draw_landmarks(img, 
                                      hand_landmarks, 
                                      mp_hands.HAND_CONNECTIONS,
                                      #mp_drawing_styles.get_default_hand_landmarks_style(),
                                      #mp_drawing_styles.get_default_hand_connections_style()
                                      )
            # limpiamos las listas
            xlist.clear()
            ylist.clear()
            # iteramso entre las cordenadas de cada punto 
            for id, lm in enumerate(hand_landmarks.landmark):
                #print(lm)
                cx, cy = int(lm.x * w), int(lm.y * h) # obtenemos x y y de cada punto y los multiplicamos por el ancho y altp
                # a침adimos a las listas
                xlist.append(cx)
                ylist.append(cy)
            # obtenemos el minimo y maximo de x que sera los puntos delimitrantes en X
            xmin, xmax = min(xlist), max(xlist)
             # obtenemos el minimo y maximo de Y que sera los puntos delimitrantes en Y
            ymin, ymax = min(ylist), max(ylist)
            
            coord = [xmin, xmax, ymin, ymax]
            # dibujamos el rectangulo delimitador de la mano 
            cv.rectangle(img, (xmin-20, ymin-20), (xmax+20, ymax+20),(0, 255, 0), 2)
    # cambiamos el espacio de color a BGR espacio en el que trabaja opencv
    img = cv.cvtColor(img, cv.COLOR_RGB2BGR)
    # retornamos la imagen y las cordenadas
    return img, coord
## *****************************************************************************              
## *****************************************************************************
## ----------- CODIGO PRINCIPAL--------------------------
## creacion del Objeto captura de video 
cap = cv.VideoCapture(0)
# si la camara se abrio correctamente
while cap.isOpened():
    # leemos el frame de la camara
    ret, frame = cap.read() 
    # si el frame se a leido correctamente
    if ret:
        #realizamos un espejado de la imagen
        frame = cv.flip(frame, 1)
        #creamos una copia de la imagen 
        img_proces = frame.copy()
        image, coord = hand_detect(img_proces) # enviamos la copia de la imagen a la funcion
        cv.imshow("MediaPipe Hands", image) # mostramso en una ventana
        
        key = cv.waitKey(10) # capturamos la entrada por teclado
        if key == 27: # si es escape salimos del bucle
            break
        elif key == ord('q'): # si es la q capturamos una imagen 
            
            # extraemos las coordenadas delimitadoras de la mano
            xmin, xmax, ymin, ymax = coord[0], coord[1], coord[2], coord[3] # 
            # realizamos un proces de rebanado para extraer solo la mano 
            img_hand = frame[ymin-20:ymax+20, xmin-20:xmax+20]
            # realizamos un cambio de dimension a (300,300,3)
            img_hand = cv.resize(img_hand, (300,300))
            # guardamos la imagen de la mano 
            cv.imwrite("./Datasets/a/a_"+str(cont)+".jpg", img_hand)
            # aumentamos en 1 el contador 
            cont +=1
            print("guardada con exito!!")
        
    else:
        break
# liberamos la camara         
cap.release()
# destruimos todas las ventanas
cv.destroyAllWindows()
```
